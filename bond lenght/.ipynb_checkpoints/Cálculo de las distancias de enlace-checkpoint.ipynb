```python
%matplotlib inline
import __main__
__main__.pymol_argv = ['pymol','-qc'] # Pymol: quiet and no GUI
import sys, pymol
from pymol import cmd, stored
stdout = sys.stdout
pymol.finish_launching()
sys.stdout = stdout 
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
import re
```

Calcular las distancias de enlace C $\alpha$ - H $\alpha$, C $\alpha$ - N y C $\alpha$ - C $\beta$, para una asamblea de ubiquitinas.

```python
job= "Ubiquitin"
sele = "N"
atom_name= "N"
pdb_file_name ='2k39.pdb'
cmd.load(pdb_file_name )
n_states = cmd.count_states('all')
n_residues = cmd.count_atoms('name CA')
distance = np.zeros((116, 76))
```

```python
for s in range(1,n_states+1):    
    conf = []
    for i in range(1 , n_residues+1):
        try:
            dst = cmd.get_distance('resi %s and name CA' % (i), 'resi %s and name %s' % (i, sele), state = s)
            conf.append(dst)
        except:
            conf.append(np.nan)
            pass
    distance[s-1] = conf
```

```python
dst_df = pd.DataFrame(distance)
```

```python
a = dst_df.dropna(1 , 'all')
```

```python
b = np.array(a)
```

```python
b[:,1:75]
```

```python
ax = sns.tsplot(data = b[:,1:75] , interpolate=False, err_style="ci_bars", ci=50)
plt.xlabel("Residue")

if sele == "HA":
    plt.ylabel("C$\\alpha$ - H$\\alpha$ Distance ($\\AA$)" )
    
elif sele == "C":
    plt.ylabel("C$\\alpha$ - C' Distance ($\\AA$)") 
    
else:
    plt.ylabel(("C$\\alpha$ - %s Distance ($\\AA$)") % (sele) )
    
plt.savefig("distances_%s_%s.png" % (job, sele), dpi = 300)
```

```python
c = b[:,0:74].flatten()
mu = np.mean(c)
sigma = np.std(c)
sigma
```

```python
sns.kdeplot(c)#, gridsize=5000)
plt.plot(0, label='$\\mu$ = {:.3f}\n$\\sigma$ = {:.3f}'.format(mu, sigma), alpha=0)
plt.xlim(1.4, 1.545)
plt.xlabel("C$\\alpha$ - %s Distance ($\\AA$)" % (atom_name))
plt.legend(loc=1, fontsize=14)
plt.savefig("distances_kde_%s_%s.png" % (job, sele), dpi = 300)
```
